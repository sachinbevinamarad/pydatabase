```python
from pandas import DataFrame
df2 = DataFrame(
               [
                ['Aryan','Singh','Full-time','Administration',2,20000],
                ['Rohan','Agarwal','Intern','Technical',3,5000],
                   ['Riya','Shah','Full-time','Administration',5,100000],
                   ['Yash','Bhatia','Part-time','Technical',7,100000],
                   ['Siddhant','Khanna','Full-time','Management',6,200000]
               ]
               ,columns=['First Name','last Name','Type','Department','YoE','Salary']
              )
df2
```

```python
import pandas as pd
table = pd.pivot_table(df2,index='Type')
table
```

```python
import pandas as pd
table = pd.pivot_table(df2,index='Type',columns='Department')
table
```

```python
import pandas as pd
table = pd.pivot_table(df2,index='Type',columns='Department',values='Salary')
table
```

```python
import pandas as pd
table = pd.pivot_table(df2,index='Type',columns='Department',values='Salary',aggfunc='sum')
table
```
