```python
survived_count= df2.groupby('Survived')['Survived'].count() #This can beused to group large amounts of data and compute operations on thesegroups.
survived_count
```

```python
pclass_count = df2.groupby('Pclass')['Pclass'].count()
pclass_count
```

```python
sex_count = df2.groupby('Sex')['Sex'].count()
sex_count
```

```python
embarked_count = df2.groupby('Embarked')['Embarked'].count()
embarked_count
```
