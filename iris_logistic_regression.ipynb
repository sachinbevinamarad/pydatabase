```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x = df.iloc[:,:3]
```

```python
y = df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=2)
```

```python
# tr = ColumnTransformer( transformers = [
#     ('t1',LabelEncoder(),['species'])
# ],remainder='passthrough')
le = LabelEncoder()
y_tr = le.fit_transform(y_train)
y_te = le.fit_transform(y_test)
y_te
```

```python
log = LogisticRegression(random_state=0)
```

```python
log.fit(x_train , y_tr)
```

```python
y_pred = log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("Accuracy Score",accuracy_score(y_te,y_pred))
```

```python
# pd.DataFrame(confusion_matrix(y_te , y_pred),columns=list(range(2)))
confusion_matrix(y_te , y_pred)
```

```python
result = pd.DataFrame()
result['Actual Label'] = y_te
result['Decision Tree Prediction'] = y_pred
result
```

```python
from sklearn.metrics import precision_score
precision_score(y_te, y_pred , average =None)
```

```python
from sklearn.metrics import recall_score
recall_score(y_te, y_pred , average = 'micro')
```

```python
from sklearn.metrics import f1_score
f1_score(y_te, y_pred , average = 'micro')
```
