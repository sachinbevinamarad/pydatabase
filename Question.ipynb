```python
from pandas import DataFrame
df = DataFrame(
               [
                ['Aryan','Singh','Full-time','Administration',2,20000],
                ['Rohan','Agarwal','Intern','Technical',3,5000],
                   ['Riya','Shah','Full-time','Administration',5,100000],
                   ['Yash','Bhatia','Part-time','Technical',7,100000],
                   ['Siddhant','Khanna','Full-time','Management',6,200000]
               ]
               ,columns=['First Name','last Name','Type','Department','YoE','Salary']
              )
df
```

```python
df[["Type","Salary"]]
```

1. Make a pivot table which shows average salary of each type of employee for each 
department.

```python
import pandas as pd
table = pd.pivot_table(df,index='Type',columns='Department',values='Salary',aggfunc='mean')
table
```

2.Make a pivot table which shows the sum and mean of the salaries of each type of 
employee and the number of employees of each type.

```python
import pandas as pd
table = pd.pivot_table(df,index='Type',values='Salary',aggfunc=['sum','mean','count'])
table
```
